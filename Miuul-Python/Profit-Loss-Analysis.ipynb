```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_ = pd.read_excel("Book2.xlsx", sheet_name="profit-loss")
df = df_.copy()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
```

```python
df.head()
```

```python
df1 = df[df["Profit"]<0]
```

```python
df1.tail()
```

```python
df1.groupby("Segment").agg({"State":"count"})
```

```python
df1.groupby("State").agg({"Profit":"count"}).sort_values(by="Profit", ascending=False)
```

```python
df1.groupby("Order_Date").agg({"Profit":"count"}).sort_values(by="Profit", ascending=False)
```

```python
df1.groupby("Customer_Name").agg({"Profit":"sum"}).sort_values(by="Profit", ascending=True)
```

```python
df1.groupby("Ship_Mode").agg({"Profit":"sum"}).sort_values(by="Profit", ascending=True)
```
