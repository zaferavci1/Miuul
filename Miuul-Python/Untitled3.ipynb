```python
df["customer_value_total_ever_offline"] = (
    df["customer_value_total_ever_offline"]
    .str.replace(",", ".", regex=False)  # Virgülü ondalık ayırıcı yap
    .astype(float)  # Float'a dönüştür
)
```

```python
df["customer_value_total_ever_online"] = (
    df["customer_value_total_ever_online"]
    .str.replace(",", ".", regex=False)  # Virgülü ondalık ayırıcı yap
    .astype(float)  # Float'a dönüştür
)
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['order_num_total_ever_online'], color="skyblue", label="Online Sipariş Sayısı", kde=True)
sns.histplot(df['order_num_total_ever_offline'], color="salmon", label="Offline Sipariş Sayısı", kde=True)
plt.legend()
plt.title("Online ve Offline Sipariş Sayısı Dağılımı")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['customer_value_total_ever_online'], color="skyblue", label="Online Sipariş Değeri", kde=True)
sns.histplot(df['customer_value_total_ever_offline'], color="salmon", label="Offline Sipariş Değeri", kde=True)
plt.legend()
plt.title("Online ve Offline Sipariş Değeri Dağılımı")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="order_channel", palette="viridis")
plt.title("Sipariş Kanallarının Kullanım Sayısı")
plt.xticks(rotation=45)
plt.show()
```

```python
# Tarih formatına çevirme (eğer değilse)
df['first_order_date'] = pd.to_datetime(df['first_order_date'])
df['last_order_date'] = pd.to_datetime(df['last_order_date'])

# Aylık sipariş sayısı
df.set_index('first_order_date').resample('M').size().plot(title="Aylık İlk Sipariş Sayısı", figsize=(12,6))
plt.show()

df.set_index('last_order_date').resample('M').size().plot(title="Aylık Son Sipariş Sayısı", figsize=(12,6))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="order_num_total", y="customer_value_total", alpha=0.6)
plt.title("Toplam Sipariş Sayısı ve Müşteri Değeri Dağılımı")
plt.xlabel("Toplam Sipariş Sayısı")
plt.ylabel("Müşteri Değeri")
plt.show()
```
