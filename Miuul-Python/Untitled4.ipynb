```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_ = pd.read_csv("archive/superStoreDB.csv")
df = df_.copy()
```

```python
df.head()
```

```python
df.describe()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
df['Sales'] = df['Sales'].replace(',', '.', regex=True).astype(float)
```

```python
df.dtypes
```

```python
df.set_index('Order_Date').resample('W').sum()['Sales']
```

```python
df.head()
```

```python
df["Order_Date_Year"]= df['Order_Date'].dt.isocalendar().year
```

```python
df["Order_Date_Week"]= df['Order_Date'].dt.isocalendar().week
```

```python
df.head()
```

```python
df1 = df.groupby(["Order_Date_Year","Order_Date_Week"]).agg({"Sales":"sum"}).reset_index()
```

```python
df1 = df1[df1["Order_Date_Year"] > 2014].reset_index()
```

```python
df1 = df1.drop("index", axis = 1)
```
